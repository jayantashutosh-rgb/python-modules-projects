Q1. What are data structures, and why are they important?

Explanation:
Data structures are methods of organizing and storing data in a computer so that operations such as searching, insertion, deletion, and updating can be performed efficiently. Choosing the right data structure improves program performance, reduces memory usage, and makes the code easier to maintain.

Example Code:

numbers = [10, 20, 30]
numbers.append(40)
print(numbers)

Q2. Explain the difference between mutable and immutable data types with examples.

Explanation:
Mutable data types can be modified after creation, such as lists, sets, and dictionaries.
Immutable data types cannot be changed once created, such as strings, tuples, and integers.

Example Code:

# Mutable
lst = [1, 2, 3]
lst.append(4)
print(lst)

# Immutable
s = "hello"
# s[0] = "H"  # This would raise an error

Q3. What are the main differences between lists and tuples in Python?

Explanation:
Lists are mutable and allow elements to be added, removed, or updated. They are written using square brackets [ ].
Tuples are immutable and written using parentheses ( ).
Tuples are usually faster and safer for fixed collections of data.

Example Code:

my_list = [1, 2, 3]
my_list[0] = 99

my_tuple = (1, 2, 3)
# my_tuple[0] = 99  # Error because tuples are immutable

Q4. Describe how dictionaries store data.

Explanation:
Dictionaries store data in the form of key–value pairs. Internally, Python uses a hash table to convert keys into indexes, allowing very fast data retrieval.

Example Code:

student = {"name": "Ashutosh Jayant", "age": 22}
print(student["name"])

Q5. Why might you use a set instead of a list in Python?

Explanation:
Sets automatically remove duplicate elements and provide faster membership checking compared to lists, making them useful when uniqueness matters.

Example Code:

nums = [1, 2, 2, 3]
print(set(nums))

Q6. What is a string in Python, and how is it different from a list?

Explanation:
A string is an immutable sequence of characters used to store text.
A list is a mutable collection that can store elements of different types.

Example Code:

s = "hello"
lst = ["h", "e", "l", "l", "o"]

lst[0] = "H"

Q7. How do tuples ensure data integrity in Python?

Explanation:
Tuples are immutable, which means their contents cannot be changed after creation. This prevents accidental modification of important data and helps preserve data integrity.

Example Code:

coordinates = (10, 20)
print(coordinates)

Q8. What is a hash table, and how does it relate to dictionaries in Python?

Explanation:
A hash table is a data structure that maps keys to values using a hashing function. Python dictionaries are implemented using hash tables, which allows them to perform fast lookups.

Example Code:

phone_book = {"Ashutosh Jayant": 9876543210}
print(phone_book["Ashutosh Jayant"])

Q9. Can lists contain different data types in Python?

Explanation:
Yes, Python lists can contain elements of different data types, such as integers, strings, floats, and even other objects.

Example Code:

mixed = [1, "apple", 3.5]
print(mixed)

Q10. Explain why strings are immutable in Python.

Explanation:
Strings are immutable for reasons of performance, memory optimization, and security. When a string is modified, Python actually creates a new object instead of changing the original one.

Example Code:

a = "hi"
b = a + "!"
print(b)


Q11. What advantages do dictionaries offer over lists for certain tasks?

Explanation:
Dictionaries allow very fast lookup using keys instead of numeric indexes. They are ideal when data is associated with unique identifiers, such as storing student names with marks. They also improve readability because keys describe the data.

Example Code:

marks = {"Ashutosh Jayant": 90, "Ravi": 85}
print(marks["Ashutosh Jayant"])

Q12. How do sets handle duplicate values in Python?

Explanation:
Sets automatically remove duplicate elements. If the same value is added multiple times, only one copy remains in the set.

Example Code:

values = {1, 2, 2, 3, 3}
print(values)

Q13. Describe a scenario where using a tuple would be preferable over a list.

Explanation:
Tuples are preferred when the data should not change, such as storing geographic coordinates, RGB color values, or database records.

Example Code:

location = (28.61, 77.20)
print(location)

Q14. How does the “in” keyword work differently for lists and dictionaries?

Explanation:
In a list, the in keyword checks whether a value exists in the list.
In a dictionary, it checks whether a key exists.

Example Code:

lst = [10, 20, 30]
print(20 in lst)

d = {"a": 1, "b": 2}
print("a" in d)

Q15. Can you modify the elements of a tuple? Explain why or why not.

Explanation:
No, elements of a tuple cannot be modified because tuples are immutable. This design helps protect data from accidental changes.

Example Code:

t = (1, 2, 3)
# t[0] = 99  # This will raise an error

Q16. What is a nested dictionary, and give an example of its use case?

Explanation:
A nested dictionary is a dictionary that contains another dictionary as one of its values. It is commonly used to represent structured data such as student records or employee details.

Example Code:

students = {
    "Ashutosh Jayant": {"math": 90, "science": 88},
    "Ravi": {"math": 85, "science": 80}
}
print(students["Ashutosh Jayant"]["math"])

Q17. Describe the time complexity of accessing elements in a dictionary.

Explanation:
On average, dictionary access takes O(1) time due to hashing. This means the value can be retrieved almost instantly, regardless of dictionary size.

Example Code:

data = {"x": 100, "y": 200}
print(data["x"])

Q18. In what situations are lists preferred over dictionaries?

Explanation:
Lists are preferred when the order of elements matters, when data is accessed using numeric indexes, or when storing a simple sequence without key–value relationships.

Example Code:

items = ["apple", "banana", "cherry"]
print(items[1])

Q19. Why are dictionaries considered unordered, and how does that affect data retrieval?

Explanation:
Dictionaries are conceptually unordered because elements are accessed by keys rather than positions. Retrieval depends on the key, not the insertion order.

Example Code:

info = {"name": "Ashutosh Jayant", "age": 22}
print(info["age"])

Q20. Explain the difference between a list and a dictionary in terms of data retrieval.

Explanation:
Lists retrieve elements using numeric indexes.
Dictionaries retrieve values using keys.
Dictionary lookup is generally faster for large datasets.

Example Code:

lst = ["red", "green", "blue"]
print(lst[1])

d = {"red": "#ff0000", "green": "#00ff00"}
print(d["green"])
